# Working with data packages (DP/dp)
## Introduction
A `dp` is a `.json` file containg the data contextual information: `location` (local path) or `URI` (remote), the `schema` (dataset structure) and its `metadata` (data about the dataset). The `dp` allows us to organise out data in a compact way, which is consisted with the [FAIR](https://www.nature.com/articles/sdata201618) principles of data stewardship and use. It promote data openness and research reproducibility. There are two broad approaches for dealing with `dp`:
- Web-based graphical app
- Programming interfaces (different languages)

For this case will look at the `Python` interface for working with `dp`. You may look at [datapackage-py](https://github.com/frictionlessdata/datapackage-py) GitHub repository for detailed instructions.

## Creating a data package
The Python `datapackage` is available for this purpose and can be installed using the pip Python installation manager using the command `pip install datapackage` from the Linux/MAC terminal. I presime you have Python already installed in your computer.


```python
from datapackage import Package
# Inferring dp from tabular data
package = Package()# Init an empty dp
package2 = Package()
package.infer('**/*.csv')# Infers a dp from all csv datasets
package2.infer('**/SurgicalDataFrame.csv')
package.descriptor# Displays the dp
#Save dps
package.save('data_dp.json')
package2.save('SurgicalDataFrame_dp.json')
# Creating a dp from scratch
package3 = Package()# Init an empty dp
# Use the descriptor attribute to assign attributes of the dp
package3.descriptor['name'] = 'period-table'
package3.descriptor['title'] = 'Periodic Table'
package3.descriptor['resources'] = 'data.csv'
package3.descriptor# Displays the dp
package3.save('arbitrary_dp.json')

package.descriptor['resources'][0]# Get first resource

package.descriptor['resources'][1]# Get second resource
# Loading an existing dp
package_ld = Package('datapackage-hrma_validation.json')
package_ld.descriptor
```
